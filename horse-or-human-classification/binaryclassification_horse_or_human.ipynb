```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D   
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Flatten
```

```python
model = Sequential()
```

```python
model.add(Conv2D(16 ,(3,3), 
                 activation='relu', 
                 input_shape=(150, 150, 3)))
```

```python
model.add(MaxPooling2D((2,2)))
```

```python
model.add(Conv2D(32, (3,3), 
                 activation='relu'))
```

```python
model.add(MaxPooling2D(2,2))
```

```python
model.add(Conv2D(64, (3,3), 
                 activation='relu'))
```

```python
model.add(MaxPooling2D(2,2))
```

```python
model.add(Flatten())
```

```python
model.add(Dense(512, activation='relu'))
```

```python
model.add(Dense(1, activation='sigmoid'))
```

```python
model.summary()
```

```python
from tensorflow.keras.optimizers import RMSprop
```

```python
model.compile(loss='binary_crossentropy', 
              optimizer=RMSprop(learning_rate=0.001),
              metrics=['acc'])
```

```python
# data Processing
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# tăng cường data, biến tất cả các phần tử trong matrix về lớn hơn 0 nhỏ hơn 1
# giông với training_imges = train_imges.adtype("float")/ 255.0
train_generator = ImageDataGenerator(rescale=1/255)
```

```python
valid_generator = ImageDataGenerator(rescale=1/255)
```

```python
# load ảnh tử folder vào thẳng generated_train_data
generated_train_data = train_generator.flow_from_directory(
    train_folder,
    target_size=(150, 150),
    batch_size=128,
    class_mode='binary'
)
```

```python
generated_valid_data = valid_generator.flow_from_directory(
    valid_folder,
    target_size=(150, 150),
    class_mode='binary'
)
```

```python
model.fit(
    generated_train_data,
    steps_per_epoch = 8,
    epochs = 15,
    verbose = 1,
    validation_data = generated_valid_data
)
```

```python
model.save("horse-or-human.h5")
```
